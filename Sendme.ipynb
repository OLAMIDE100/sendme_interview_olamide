```python
import pandas as pd
import re
```

```python
data = pd.read_excel(r"Raw Data.xlsx",header=None)
```

```python
data.head(20)
```

```python
def startsandendswith(main_str):
    return main_str.startswith("*") and main_str.endswith("*")
```

```python
l = ['message','Done','Transfer','(Minus','Cancelled','Cash','*20/05/21*','No']
big_list = []
small_list = []

for row in data[0]:
    if isinstance(row,int):
        small_list.append(row)
    elif any(map(lambda x:x in l,row.split())):
       pass
    elif not startsandendswith(row):
        small_list.append(row)
    else:
        small_list.append(row)
        big_list.append(small_list)
        small_list = []
    
        
```

```python
big_list
```

```python
new_big_list = []
for k in big_list:
    if 'EDITED' not in k[0] and 'Edited' not in k[0]:
        new_big_list.append(k)
        
new_big_list
```

```python
def contact(row):
    no = str(row['CONTACT'])
    if no[0] == '2':
        return "+" + no
    else:
        return "+234" + no
```

```python
delivery_time = []
ordered_time = []
amount = []
location = []
phone_no = []
name = []
date = []
item = []

for p in new_big_list:
    delivery_time.append(re.sub('[*]+', '', p[-1]))
    ordered_time.append(p[0].split(' ')[1])
    amount.append(re.sub("[^\d\.]", "", p[-2]))
    location.append(p[-3])
    phone_no.append(p[-4])
    name.append(p[-5])
    date.append(re.sub('[,]+', '', p[0].split(' ')[0]))
    item.append(','.join(p[1:-5]))

```

```python
len(item)
```

```python
df = pd.DataFrame(list(zip(date, ordered_time,item,location,name,phone_no,amount,delivery_time)),
               columns =['DATE','ORDER TIME','ITEM','ADDRESS','NAME','CONTACT','AMOUNT','DELIVERY TIME'])
```

```python
df['CONTACT'] = df.apply(lambda row: contact(row), axis=1)
```

```python
with pd.ExcelWriter('tabular_sendme_data.xlsx') as writer:
            df.to_excel(writer, sheet_name = 'sheet1', index = True)
```
